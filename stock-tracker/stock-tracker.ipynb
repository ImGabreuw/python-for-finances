```python
import yfinance as yf
import matplotlib.pyplot as plt
```

# Selection of portfolio stocks

```python
stocks = ["AAPL", "MSFT", "PETR4.SA", "VALE3.SA"]
```

# Import financial assets data

```python
stocks_data = yf.download(stocks, start="2020-01-01")
```

```python
stocks_data["Adj Close"]
```

```python
stocks_data["Adj Close"].plot()

plt.title("Stock tracker")
plt.xlabel("Date")
plt.ylabel("Adjusted price")
plt.legend(fontsize=7);
```

# Price normalization (data scaling)

```python
normalized_data = stocks_data["Adj Close"] / stocks_data["Adj Close"].iloc[0]
```

```python
normalized_data.plot()

plt.title("Stock Tracker (normalized)")
plt.xlabel("Date")
plt.ylabel("Ajusted price")
plt.legend(fontsize=7);
```
