```python
import yfinance as yf
import vectorbt as vbt
import pandas as pd
```

# Assets selection

```python
assets = ["BBAS3.SA", "B3SA3.SA", "ITSA4.SA", "PETR4.SA", "VALE3.SA",
          "TRPL4.SA", "EGIE3.SA", "TAEE11.SA", "WEGE3.SA", "KLBN4.SA"]
```

# Asset data extraction

```python
start = "2018-01-01"
```

```python
dividends = vbt.YFData.download(assets, start=start).get("Dividends")
```

```python
dividends
```

```python
close = yf.download(assets, start=start)["Close"]
```

# Calculation of dividends in the last 12 months

```python
# The argument "365D" represents exacly 1 year for Pandas (only works if index is of datatime type)
dividends_last_12m = dividends.rolling('365D').sum()
```

```python
dividends_last_12m.tail()
```

# Calculation of annual dividend yield

```python
dividends_last_12m
```

```python
close
```

```python
dividend_yield_last_12m = (dividends_last_12m / close) * 100
```

```python
dividend_yield_last_12m = dividend_yield_last_12m.dropna()
dividend_yield_last_12m
```

```python
dividend_yield_last_12m.vbt.plot(width=1000, height=600).show()
```
